Average GPA

```python
average_gpa = data['CGPA'].mean()
print("Average GPA:", "{:.2f}".format(average_gpa))

cgpa_data = data['CGPA']
cgpa_filtered = cgpa_data[(cgpa_data >= 1) & (cgpa_data <= 10)]

# Define bins for CGPA ranges
bins = [1, 3, 6, 8, 10]  # Upper bound of 11 to include 10 in the last bin
labels = ['1-3', '3-6', '6-8', '8-10']

# Categorize the CGPA data into bins
cgpa_binned = pd.cut(cgpa_filtered, bins=bins, labels=labels, right=False, include_lowest=True)

# Count the number of occurrences in each bin
cgpa_counts = cgpa_binned.value_counts().sort_index()

# Values for the donut chart
values = cgpa_counts.values
labels = cgpa_counts.index

# Filter out values with 0 occurrences
non_zero_indices = values != 0
values = values[non_zero_indices]
labels = labels[non_zero_indices]

# Create the donut chart
fig, ax = plt.subplots()
ax.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired(range(len(labels))))
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')
plt.title('CGPA Distribution Donut Chart')
plt.show()
```

```python
import pandas as pd

# Define a function to convert income to numeric values
def convert_income(income):
    if isinstance(income, str):
        income = income.replace('Lakh', '00000')
        income = income.replace('+', '')
        income = income.replace(' ', '')  # Remove any extra spaces
        if '-' in income:
            low, high = income.split('-')
            return (float(low) + float(high)) / 2
        else:
            return float(income)
    return income

# Apply the conversion function to the 'Family_Income' column
data['Family Income'] = data['Family Income'].apply(convert_income)

# Verify the conversion
print(data['Family Income'].head())

# Now you can calculate the correlation and visualize the data
correlation = data['Family Income'].corr(data['CGPA'])
print(f"Correlation between family income and GPA: {correlation:.2f}")

# Visualization of the relationship
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='Family Income', y='CGPA', data=data)
plt.title('Relationship between Family Income and GPA')
plt.xlabel('Family Income')
plt.ylabel('GPA')
plt.show()
```
